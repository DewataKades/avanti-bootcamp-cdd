```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

# Chamando e lendo o dataframe

```python
df=pd.read_csv('http://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

## Ordenando e plotando os dados

```python
#valores simples
df=df.sort_values(by='arrecadacao', ascending=False)
df=df.assign(arrecadacao_milhao= (df.arrecadacao)/(10**6))
sns.set_style("whitegrid")
plt.figure(figsize=(5,2))
ax=sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')
ax.bar_label(
    ax.containers[0],
    fmt='%0.1f',
    label_type='center',
    color='white')
plt.xlabel('Arrecadação (Milhões de dolares)')
plt.ylabel('Filme')
plt.title('Arrecadação dos filmes')
ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)
plt.show()
```
